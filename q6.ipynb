## Question 6

Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt
us = pd.read_csv('us-marriages-divorces-1867-2014.csv', error_bad_lines=False)
mrg=us['Marriages_per_1000'].tolist()
div=us['Divorces_per_1000'].tolist()
year=us['Year'].tolist()
plt.plot(year, mrg,label = "Marriages")
plt.plot(year,div,label = "Divorces")
plt.xlabel('Year')
plt.ylabel('marriages and divorces per capita')
plt.title(' marriages and divorces per capita in the U.S. between 1867 and 2014.')
plt.legend()
plt.show()
```
